<code>enumerate() funksiyası</code>

funksiya vasitəsilə verilən tipi elementləri sırası ilə nömrələndirmək mümkündür.Nömrələməni kortej tipində edərək bizə enumerate dəyər göndərir.

```python
def example(y):
    sayğac =0
    siyahı = []
    for i in y:
        sayğac+=1
        siyahı.append((sayğac,i))
    return siyahı

example(['Python','Java','C++']) 
```

Yuxarıda funksiyada qeyd etdiyimiz elementləri nömrələndirdik.Və ilk 1-ədədindən başladı.Yuxarıdakı funksiya əvəzinə<code> enumerate </code>funksiyasını istifadə edərək qısa yoldan nəticəni əldə edə bilərik

```python
siyahı = ['Python','Java','C++','Perl']
print(list(enumerate(siyahı)))#enumerate dəyərini siyahı tipinə çevirdiyimiz üçün siyahı tipində məlumat alacağıq
print(type(siyahı)) 
```

```python
a = [10, 20, 30, 40]
for i in enumerate(a):
    print(i)
```

```python
dictionary = {1: 'a', 2: 'b', 3: 'c'}
for i in enumerate(dictionary.values()):
    print(i)
```

<code>zip() funksiyası</code>

Funksiya parametr olaraq aldığı elementləri qruplandıraraq dəyərini kortej tipində bizə göndərir.İki verilən tipi elementləri arasında say etibarı ilə bərabərlik ödəndiyi iterasiyaya qədər davam edər.Qruplandıracağı qarşılığı olmadıqda isə iterasiya sonlandırılır.

```python
def Group(iter1,iter2):
    sayğac = 0
    siyahı = []
    while (sayğac<len(iter1) and sayğac<len(iter2)):
        siyahı.append((iter1[sayğac],iter2[sayğac]))
        sayğac+=1
    return siyahı
    
Group(['Azerbaijan','Germany','Greek'],['Baki','Berlin','Athena','Tehran'])
        
```

Yuxarıdakı funksiya vasitəsilə hər iki siyahı daxilindəki elementləri qruplaşdırdıq.Tehran elementinə qarşılıq element olmadığından ifadə nəzərə alınmadı

```python
var = ['Azerbaijan','Germany','Greek']
var2 = ['Baki','Berlin','Athena']
for i in list(zip(var,var2)):
    print(i)
```

```python
#və ya
var = ['Azerbaijan','Germany','Greek']
var2 = ['Baki','Berlin','Athena']
var3 = [12,16,18,20,22,24,26]
print(list(zip(var,var2,var3))) #zip() funksiyası üçün verilən tipi məhdudiyyəti yoxdur
```

<code> filter() funksiyası</code>

Funksiya bool verilən tipində dəyərləndirir.True və ya False.Həmçinin parametr olaraq funksiya və verilən tipində ifadələri alır.İfadələri funksiyaya göndərərək True elementlərini bizə göstərir.Misal olaraq tək və cüt ədədlərdən ibarət diapazonda rəqəmlər var.Biz süzgəcləndirərək sadəcə cüt ədədləri ekrana çap etdirə bilərik

```python
siyahı = []
for i in range(20):
    if i%2==0:
        siyahı.append(i)
print(siyahı)
        
```

```python
def func(i):
    if i%2 == 0:
        return i
var =range(20)
list(filter(func,var))        
```
